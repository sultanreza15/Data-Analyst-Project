# **Latar Belakang**

FundFusion, sebuah institusi bank di Wakanda, ingin mengembangkan upaya pelayanan marketingnya terhadap nasabah existing maupun nasabah baru. Salah satu strategi marketing-nya adalah melakukan kampanye berbasis telepon atau telemarketing. Cara kerja strategi tersebut adalah menggunakan media telepon untuk menawarkan produk kepada nasabah.

Saat ini, FundFusion belum memiliki strategi yang jelas dalam hal penawaran produk ketika menghubungi calon nasabah. Masing-masing account manager memiliki KPI yang berbeda untuk customer acquisition berdasarkan produk yang ditawarkan oleh FundFusion, seperti Tabungan, Deposito, Kartu Kredit, Kredit Rumah, Kredit Kendaraan, maupun Kredit Dana Tunai.

Hal ini mengakibatkan calon nasabah sehari dapat dihubungi berulang kali oleh account manager yang berbeda untuk ditawarkan produk yang berbeda pula. Implikasinya, rejection rate yang didapatkan sangat tinggi dan dianggap sebagai spam oleh calon nasabah yang berujung pada pemblokiran nomor.

# **Problem Statement**

Belum memiliki strategi yang tepat untuk menawarkan jenis produk yang sesuai dengan segmen calon nasabah yang akan direkrut

# **Objective**

Membuat Sebuah Model Clustering untuk mengetahui kepemilikan produk berdasarkan demografi nasabah yang saat ini sudah menggunakan layanan FundFusion dengan Silhouette Score >0.7

## Pengecekan Korelasi

```python
corrtest1 = data1.corr().abs()
corrtest2 = X_num.corr().abs()
```

```python
# Membuang Nilai Redundan Pada Matriks
upper = corrtest1.where(np.triu(np.ones(corrtest1.shape), k=1).astype(np.bool))
upper1 = corrtest2.where(np.triu(np.ones(corrtest2.shape), k=1).astype(np.bool))

# Mencari Nilai Yang Berkorelasi di atas 0.7
to_drop = [column for column in upper.columns if any(upper[column] > 0.7)]
to_drop1 = [column for column in upper1.columns if any(upper1[column] > 0.7)]

# Menghapus Kolom Yang Berkorelasi di atas 0.7
data1 = data1.drop(to_drop, axis=1)
data2 = data2.drop(to_drop1, axis=1)
```

# **Modeling & Evaluation**


Pembangunan model akan menggunakan 2 algoritma:

## K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Experiment 0

```python
# Variasi hyperparameter (jumlah cluster)
for n_clusters in range(3, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(data_combined)
    silhouette_avg = silhouette_score(data_combined, cluster_labels)
    print(f"Silhouette Score (K-Means) -"+str(n_clusters)+" : "+str(silhouette_avg))
    data0['Clustering_KMeans_Exp0_'+str(n_clusters)] = cluster_labels

```

# **Kesimpulan**

Berdasarkan hasil pemodelan clustering menggunakan algoritma K-Means, nasabah FundFusion secara alami terbagi menjadi 3 cluster utama yang memiliki karakteristik demografi dan perilaku finansial yang sangat berbeda.

Ketiga cluster ini menjawab permasalahan tingginya rejection rate akibat penawaran produk yang tidak relevan (spamming). Berikut adalah profil ketiga cluster nasabah:

1. Cluster Muda Potensial
- Profil: Rata-rata usia 29 tahun.
- Karakteristik Unik: Ini adalah kelompok nasabah yang paling melek produk dasar. Terlihat dari kepemilikan Produk Tabungan yang mencapai 100%.
- Insight: Mereka masih di tahap awal membangun aset, namun sudah memiliki Total Relationship Balance yang sehat. Mereka adalah target jangka panjang yang ideal.
- Penawaran yang tidak direkomendasikan: Tabungan.
- Penawaran yang direkomendasikan: Kredit Kendaraan & Kartu Kredit.
- Alasan: 100% dari mereka sudah punya tabungan. Jangan buang waktu menawarkan hal yang sudah mereka miliki. Tawarkan produk gaya hidup (CC) atau kendaraan pertama (KKB) yang relevan dengan usia 20-an.

2. Cluster Keluarga Mapan
- Profil: Rata-rata usia 45 tahun.
- Karakteristik Unik: Berada di fase hidup yang stabil dengan kepemilikan produk yang merata (hampir 98% punya Tabungan). Jumlah anak rata-rata 1-2 orang, mengindikasikan kebutuhan biaya pendidikan atau renovasi aset.
- Penawaran yang tidak direkomendasikan: Tabungan Dasar.
- Penawaran yang direkomendasikan: Deposito & Kredit Rumah.
- Alasan: Fokus pada produk investasi jangka menengah (Deposito) untuk dana pendidikan anak, atau top-up fasilitas kredit untuk kebutuhan rumah tangga.

3. Cluster Senior Prioritas
- Profil: Rata-rata usia 59 tahun.
- Karakteristik Unik: Menariknya, meskipun usia senior, kepemilikan Produk Tabungan di segmen ini paling rendah (hanya ~65%). Padahal, mereka memiliki balance keuangan yang tinggi. Ini adalah anomali yang menjadi peluang besar.
- Penawaran yang tidak direkomendasikan: Kredit Jangka Panjang.
- Penawaran yang direkomendasikan: Tabungan & Deposito Prioritas.
- Alasan: Ada 35% nasabah senior yang belum punya produk Tabungan di kita (peluang akuisisi low hanging fruit). Hindari menawarkan kredit tenor panjang karena faktor usia dan risiko.


Dengan menerapkan strategi clustering ini, diproyeksikan FundFusion akan mendapatkan manfaat sebagai berikut:

1. Efisiensi Telemarketing: Account Manager tidak lagi membuang waktu menghubungi nasabah muda untuk menawarkan pembukaan rekening tabungan.
2. Penurunan Churn & Blokir: Nasabah tidak lagi merasa diteror oleh penawaran yang tidak relevan, meningkatkan citra brand FundFusion.
3. Peningkatan Cross-Selling: Fokus pada penawaran Tabungan ke segmen Senior yang belum memilikinya berpotensi meningkatkan Dana Pihak Ketiga (DPK) secara signifikan.
